```python
from miscope import catalog, load_family
RENDER = True
EXPORT = True
family = load_family("modulo_addition_1layer")
views_bk = ["loss_curve", "centroid_global_pca", "centroid_dmd_eigenvalues", "centroid_dmd_residual", "centroid_dmd_reconstruction"]
views = ["centroid_global_pca", "centroid_pca", "centroid_dmd_residual", "centroid_dmd_reconstruction", "centroid_dmd_eigenvalues"]
# for each variant, select representative epochs for
#   post-train loss
#   pre-grok (pre test-loss)
#   mid-grok (mid test-loss)
#   post-grok (post test-loss)
#   final
#   Note that for 101/999, this is hard to capture. 18000 is a known special moment in mid-training
key_epoch_descriptions_bk = ["post-train loss", "pre-grok (pre test-loss)", "mid-grok (mid test-loss)", "post-grok (post test-loss)", "final"]
key_epochs_bk = {
    "modulo_addition_1layer_p113_seed999": [2000, 9000, 11500, 13500, 24999],
    "modulo_addition_1layer_p101_seed999": [2000, 12500, 18000, 24500, 34999],
    "modulo_addition_1layer_p109_seed485": [1500, 3100, 4600, 6000, 24999],
    "modulo_addition_1layer_p59_seed485": [2000, 23500, 27000, 30000, 34999],
}
key_epoch_descriptions = ["pre-grok (pre test-loss)", "mid-grok (mid test-loss)", "post-grok (post test-loss)"]
key_epochs = {
    "modulo_addition_1layer_p113_seed999": [9000, 11500, 13500],
    "modulo_addition_1layer_p101_seed999": [12500, 18000, 24500],
    "modulo_addition_1layer_p109_seed485": [3100, 4600, 6000],
    "modulo_addition_1layer_p59_seed485": [23500, 27000, 30000],
    "modulo_addition_1layer_p101_seed485": [13600, 15700, 17400],
    "modulo_addition_1layer_p107_seed485": [17000, 19000, 21000],
}
```

```python
def show_views(variant):
    variant_name = variant.name
    for view_name in views:
        if view_name in ["centroid_dmd_reconstruction"]:
            n = len(key_epoch_descriptions)
            for i in range(0, n):
                epoch = key_epochs[variant_name][i]
                epoch_label = key_epoch_descriptions[i]
                print(f"Epoch: {epoch} ({epoch_label})")
                ctx = variant.at(epoch=epoch)
                if RENDER:
                    ctx.view(view_name).show()
                if EXPORT:
                    ctx.view(view_name).export(format="png")
        elif view_name == "centroid_pca":
            checkpoints = variant.get_available_checkpoints()
            ctx = variant.at(checkpoints[-1])
            if RENDER:
                ctx.view(view_name).show()
            if EXPORT:
                ctx.view(view_name).export(format="png")
        else:
            if RENDER:
                variant.view(view_name).show()                   
            if EXPORT:
                variant.view(view_name).export(format="png")
```

```python
# Canon variant: 113/999
variant = family.get_variant(prime=113, seed=999)
print(variant.name)
show_views(variant)
```

```python
# Failed/Slow grokker: 101/999
variant = family.get_variant(prime=101, seed=999)
show_views(variant)
```

```python
# Fast grokker: 109/485
variant = family.get_variant(prime=109, seed=485)
show_views(variant)
```

```python
# Successful but impaired slow grokker: 59/485
variant = family.get_variant(prime=59, seed=485)
show_views(variant)
ctx = variant.at(34999)
if RENDER:
    ctx.view("centroid_pca").show()
if EXPORT:
    ctx.view("centroid_pca").export(format="png")
```
