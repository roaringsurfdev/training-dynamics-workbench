# PC3 Dynamics — The Third Dimension Across Sites and Scales

Models tend to organize class centroids into a 2D ring (PC1+PC2 ≈ 1.0).
But PC3 is not static — it expands and contracts with training.

**Hypothesis**: PC3 is operational. The model uses it as working space:
- MLP uses PC3 early (rising first) as it begins to organize
- Attention compresses PC3 to minimum just before grokking (ring at maximum planarity)
- Post-grokking: PC3 re-expands in the attention stream

**Two scales**:
1. **Activation-space PC3** (from `repr_geometry`) — how class centroids spread into 3D at each site
2. **Weight-space PC3** (from `parameter_snapshot` + global W_in PCA) — how frequency group centroids move in 3D neuron space

**Question**: Are these in sync? Does one scale lead the other?

```python
import sys
sys.path.insert(0, '../src')

import json
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from sklearn.decomposition import PCA

from miscope import load_family
from miscope.analysis.artifact_loader import ArtifactLoader
from miscope.analysis.library import extract_neuron_weight_matrix

RESULTS_BASE = '../results/modulo_addition_1layer'
FAMILY_NAME  = 'modulo_addition_1layer'
family = load_family(FAMILY_NAME)

# Health spectrum: fast healthy → canon healthy → late → degraded
VARIANTS = [
    {'label': 'p109/s485/ds598', 'dir': 'modulo_addition_1layer_p109_seed485_dseed598', 'grok': 5243,  'health': 'healthy', 'color': '#1f77b4', 'prime': 109, 'seed': 485, 'dseed': 598},
    {'label': 'p113/s999/ds598', 'dir': 'modulo_addition_1layer_p113_seed999_dseed598', 'grok': 12448, 'health': 'healthy', 'color': '#2ca02c', 'prime': 113, 'seed': 999, 'dseed': 598},
    {'label': 'p101/s485/ds598', 'dir': 'modulo_addition_1layer_p101_seed485_dseed598', 'grok': 16656, 'health': 'late',    'color': '#ff7f0e', 'prime': 101, 'seed': 485, 'dseed': 598},
    {'label': 'p101/s999/ds598', 'dir': 'modulo_addition_1layer_p101_seed999_dseed598', 'grok': -1,    'health': 'late',    'color': '#d62728', 'prime': 101, 'seed': 999, 'dseed': 598},
    {'label': 'p59/s485/ds598',  'dir': 'modulo_addition_1layer_p59_seed485_dseed598',  'grok': -1,    'health': 'degraded','color': '#9467bd', 'prime': 59,  'seed': 485, 'dseed': 598},
]

# Augment with second_descent_onset_epoch and effective_dimensionality_cross_over_epoch
# from variant_summary.json. These are the canonical grokking markers going forward.
#   onset    = second_descent_onset_epoch  (start of second descent — use for vlines)
#   eff_xover = effective_dimensionality_cross_over_epoch (W_out PR crosses W_in PR)
#   grok     = test_loss_threshold_first_epoch (retained for lead/lag calculations)
for v in VARIANTS:
    summary_path = f'{RESULTS_BASE}/{v["dir"]}/variant_summary.json'
    try:
        with open(summary_path) as f:
            vs = json.load(f)
        v['onset']     = vs.get('second_descent_onset_epoch', -1)
        v['eff_xover'] = vs.get('effective_dimensionality_cross_over_epoch', -1)
    except FileNotFoundError:
        v['onset']     = -1
        v['eff_xover'] = -1

SITES = ['attn_out', 'mlp_out', 'resid_post']
SITE_COLORS = {'attn_out': '#2ca02c', 'mlp_out': '#d62728', 'resid_post': '#9467bd'}
SITE_LABELS = {'attn_out': 'Attn (mid-stream)', 'mlp_out': 'MLP (output delta)', 'resid_post': 'Resid Post'}

def load_repr_geometry(variant_dir):
    loader = ArtifactLoader(f'{RESULTS_BASE}/{variant_dir}/artifacts')
    s = loader.load_summary('repr_geometry')
    return s

repr_data = {v['label']: load_repr_geometry(v['dir']) for v in VARIANTS}

print(f'{"Label":<25} {"onset":>8} {"eff_xover":>10} {"grok (thresh)":>14}')
print('-' * 60)
for v in VARIANTS:
    print(f"{v['label']:<25} {v['onset']:>8} {v['eff_xover']:>10} {v['grok']:>14}")
print('\nLoaded repr_geometry summaries.')
```

## Cell 4: mean_dim vs PC3 — Two Dimensionality Signals

Two different views of dimensionality:
- **PC3 variance fraction** — how class *centroids* spread into a 3rd direction (inter-class geometry)
- **mean_dim** — average effective dimensionality of each *class cloud* (intra-class spread)

These are complementary: PC3 captures global structure (is the ring 3D?); mean_dim captures
local cloud compactness (are each class's activations tight?).

Expected: mean_dim should crash (classes collapse to points) while PC3 may expand (the ring
becomes 3D). They measure orthogonal aspects of geometric maturation.

```python
FOCUS_VARIANT = 'p113/s999/ds598'
focus_v       = next(v for v in VARIANTS if v['label'] == FOCUS_VARIANT)
FOCUS_ONSET   = focus_v['onset']
FOCUS_EFF_XOVER = focus_v['eff_xover']

s = repr_data[FOCUS_VARIANT]
epochs = s['epochs']

fig = make_subplots(rows=len(SITES), cols=1, shared_xaxes=True, vertical_spacing=0.06,
                    subplot_titles=[SITE_LABELS[site] for site in SITES])

for row_idx, site in enumerate(SITES, 1):
    pc3     = s[f'{site}_pca_var_pc3']
    md      = s[f'{site}_mean_dim']
    color   = SITE_COLORS[site]

    fig.add_trace(go.Scatter(x=epochs, y=pc3, mode='lines', name='PC3 frac',
                             legendgroup='pc3', showlegend=(row_idx == 1),
                             line=dict(color=color, width=2.5)),
                  row=row_idx, col=1)

    md_norm = (md - md.min()) / (md.max() - md.min() + 1e-9)
    fig.add_trace(go.Scatter(x=epochs, y=md_norm, mode='lines',
                             name='mean_dim (norm)', legendgroup='mean_dim',
                             showlegend=(row_idx == 1),
                             line=dict(color='#aaaaaa', width=1.5, dash='dot')),
                  row=row_idx, col=1)

    if FOCUS_ONSET > 0:
        fig.add_vline(x=FOCUS_ONSET, line=dict(color='black', dash='dash', width=1.5),
                      row=row_idx, col=1)
    if FOCUS_EFF_XOVER > 0:
        fig.add_vline(x=FOCUS_EFF_XOVER, line=dict(color='rgba(60,60,210,0.6)', dash='dot', width=1.5),
                      row=row_idx, col=1)

    fig.update_yaxes(title_text='PC3 / norm dim', row=row_idx, col=1)

fig.update_xaxes(title_text='Epoch', row=len(SITES), col=1)
fig.update_layout(
    title=f'{FOCUS_VARIANT} — PC3 (solid) vs mean_dim normalised (dotted gray)<br>'
          f'<sup>Dashed: onset ({FOCUS_ONSET})  ·  Dotted: eff_dim_xover ({FOCUS_EFF_XOVER})</sup>',
    height=600, legend=dict(orientation='h', y=1.03),
)
fig.show()
```

## Cell 5: Weight-Space Group Centroid Trajectories

In the macro surface view (mlp_macro_surface.ipynb), we see MLP neurons as a 3D cloud
in global W_in PCA space. Here we collapse each frequency group to its centroid and
track that centroid's trajectory over training.

**What this shows**: how each frequency group's collective position in weight space evolves.
PC3 of each group centroid = how far the group has moved 'out of the ring plane' in weight space.

**Link to activation space**: if weight-space PC3 and activation-space PC3 are synchronized,
the geometry is coherent across scales.

```python
GROUP_COLORS = [
    '#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
    '#8c564b', '#e377c2', '#17becf', '#bcbd22', '#7f7f7f',
]

def load_group_centroid_trajectories(variant_dir, variant_obj) -> dict:
    """Compute per-group centroid trajectory in global W_in PCA space.

    Returns:
        epochs:      (n_epochs,)
        centroids:   (n_epochs, n_groups, 3) — group centroids in global PCA space
        group_freqs: (n_groups,) 1-indexed
        var_ratio:   (3,) PCA explained variance
    """
    loader = ArtifactLoader(f'{RESULTS_BASE}/{variant_dir}/artifacts')

    ngpca       = loader.load_cross_epoch('neuron_group_pca')
    group_idx   = ngpca['neuron_group_idx']              # (d_mlp,)
    group_freqs = (ngpca['group_freqs'] + 1).astype(int) # 1-indexed
    epochs_arr  = ngpca['epochs']                        # (n_epochs,)

    epoch_list = sorted(epochs_arr.tolist())
    final_snap = loader.load_epoch('parameter_snapshot', epoch_list[-1])
    W_in_final = extract_neuron_weight_matrix(final_snap)  # (d_model, d_mlp)

    pca = PCA(n_components=3, random_state=0)
    pca.fit(W_in_final.T)  # (d_mlp, d_model)

    n_groups = len(group_freqs)
    all_centroids = np.zeros((len(epoch_list), n_groups, 3), dtype=np.float32)

    for ep_i, epoch in enumerate(epoch_list):
        snap  = loader.load_epoch('parameter_snapshot', epoch)
        W_in  = extract_neuron_weight_matrix(snap)
        proj  = pca.transform(W_in.T)                      # (d_mlp, 3)
        for g in range(n_groups):
            mask = group_idx == g
            if mask.any():
                all_centroids[ep_i, g] = proj[mask].mean(axis=0)

    return {
        'epochs':      np.array(epoch_list),
        'centroids':   all_centroids,
        'group_freqs': group_freqs,
        'var_ratio':   pca.explained_variance_ratio_,
    }


WEIGHT_FOCUS     = 'p113/s999/ds598'
weight_focus_v   = next(v for v in VARIANTS if v['label'] == WEIGHT_FOCUS)
WEIGHT_FOCUS_ONSET    = weight_focus_v['onset']
WEIGHT_FOCUS_EFF_XOVER = weight_focus_v['eff_xover']

print(f'Loading weight-space group centroid trajectories for {WEIGHT_FOCUS}...')
weight_traj = load_group_centroid_trajectories(
    weight_focus_v['dir'],
    family.get_variant(prime=weight_focus_v['prime'], seed=weight_focus_v['seed'],
                       data_seed=weight_focus_v['dseed']),
)
print(f'  {len(weight_traj["epochs"])} epochs, {len(weight_traj["group_freqs"])} groups: {weight_traj["group_freqs"].tolist()}')
print(f'  Global W_in PCA variance: {weight_traj["var_ratio"]*100}')
print(f'  second_descent_onset={WEIGHT_FOCUS_ONSET}  eff_dim_xover={WEIGHT_FOCUS_EFF_XOVER}')
```

```python
epochs_w    = weight_traj['epochs']
centroids_w = weight_traj['centroids']   # (n_epochs, n_groups, 3)
group_freqs = weight_traj['group_freqs']
n_groups    = len(group_freqs)

pc_labels = ['PC1 (weight space)', 'PC2 (weight space)', 'PC3 (weight space)']
fig = make_subplots(rows=3, cols=1, shared_xaxes=True, vertical_spacing=0.05,
                    subplot_titles=pc_labels)

for g in range(n_groups):
    color = GROUP_COLORS[g % len(GROUP_COLORS)]
    freq  = int(group_freqs[g])
    for pc_dim in range(3):
        fig.add_trace(go.Scatter(
            x=epochs_w,
            y=centroids_w[:, g, pc_dim],
            mode='lines',
            name=f'freq {freq}',
            legendgroup=f'freq{freq}',
            showlegend=(pc_dim == 0),
            line=dict(color=color, width=2),
            hovertemplate=f'freq {freq}<br>Epoch %{{x}}<br>PC{pc_dim+1}: %{{y:.3f}}<extra></extra>',
        ), row=pc_dim + 1, col=1)

if WEIGHT_FOCUS_ONSET > 0:
    for row in [1, 2, 3]:
        fig.add_vline(x=WEIGHT_FOCUS_ONSET,
                      line=dict(color='black', dash='dash', width=1.5),
                      row=row, col=1)
if WEIGHT_FOCUS_EFF_XOVER > 0:
    for row in [1, 2, 3]:
        fig.add_vline(x=WEIGHT_FOCUS_EFF_XOVER,
                      line=dict(color='rgba(60,60,210,0.6)', dash='dot', width=1.5),
                      row=row, col=1)

for row, label in enumerate(['PC1', 'PC2', 'PC3'], 1):
    fig.update_yaxes(title_text=label, row=row, col=1)
fig.update_xaxes(title_text='Epoch', row=3, col=1)

fig.update_layout(
    title=f'{WEIGHT_FOCUS} — frequency group centroid trajectories in global W_in PCA space<br>'
          f'<sup>var: PC1={weight_traj["var_ratio"][0]:.1%} PC2={weight_traj["var_ratio"][1]:.1%} '
          f'PC3={weight_traj["var_ratio"][2]:.1%}  ·  '
          f'Dashed: onset ({WEIGHT_FOCUS_ONSET})  ·  Dotted: eff_dim_xover ({WEIGHT_FOCUS_EFF_XOVER})</sup>',
    height=700, legend=dict(orientation='h', y=1.03),
)
fig.show()
```

## Cell 6: Cross-Scale Comparison — Weight-Space PC3 vs Activation-Space PC3

The key synchrony question: does PC3 in weight space (group centroid height) move
in sync with PC3 in activation space (class centroid ring-plane budget)?

Weight-space PC3 = absolute coordinate of group centroid in global W_in PCA.
For comparison: use the abs-deviation from the mean PC3 to capture 'expansion' regardless of sign.

Activation-space PC3 = variance fraction of class centroids in 3rd principal direction.

```python
s_act = repr_data[WEIGHT_FOCUS]
epochs_act = s_act['epochs']

fig = make_subplots(
    rows=2, cols=1, shared_xaxes=True,
    vertical_spacing=0.10,
    subplot_titles=[
        'Activation-space PC3 (class centroid variance fraction) per site',
        'Weight-space |PC3| (group centroid deviation from mean) per frequency group',
    ],
)

for site in SITES:
    pc3 = s_act[f'{site}_pca_var_pc3']
    fig.add_trace(go.Scatter(
        x=epochs_act, y=pc3,
        mode='lines', name=SITE_LABELS[site],
        legendgroup=f'act_{site}',
        line=dict(color=SITE_COLORS[site], width=2.5),
    ), row=1, col=1)

pc3_vals_w = centroids_w[:, :, 2]  # (n_epochs, n_groups)
mean_pc3   = pc3_vals_w.mean(axis=1, keepdims=True)
pc3_dev    = np.abs(pc3_vals_w - mean_pc3)

for g in range(n_groups):
    color = GROUP_COLORS[g % len(GROUP_COLORS)]
    freq  = int(group_freqs[g])
    fig.add_trace(go.Scatter(
        x=epochs_w, y=pc3_dev[:, g],
        mode='lines', name=f'freq {freq} (W_in)',
        legendgroup=f'w_{g}',
        line=dict(color=color, width=2, dash='dot'),
    ), row=2, col=1)

for row in [1, 2]:
    if WEIGHT_FOCUS_ONSET > 0:
        fig.add_vline(x=WEIGHT_FOCUS_ONSET,
                      line=dict(color='black', dash='dash', width=1.5),
                      row=row, col=1)
    if WEIGHT_FOCUS_EFF_XOVER > 0:
        fig.add_vline(x=WEIGHT_FOCUS_EFF_XOVER,
                      line=dict(color='rgba(60,60,210,0.6)', dash='dot', width=1.5),
                      row=row, col=1)

fig.update_yaxes(title_text='PC3 variance fraction', row=1, col=1)
fig.update_yaxes(title_text='|PC3 deviation| (a.u.)', row=2, col=1)
fig.update_xaxes(title_text='Epoch', row=2, col=1)
fig.update_layout(
    title=f'{WEIGHT_FOCUS} — activation-space vs weight-space PC3 dynamics<br>'
          f'<sup>Dashed: onset ({WEIGHT_FOCUS_ONSET})  ·  Dotted: eff_dim_xover ({WEIGHT_FOCUS_EFF_XOVER})</sup>',
    height=700, legend=dict(orientation='h', y=1.03),
)
fig.show()
```

## Cell 7: 3D Group Centroid Trajectories (Spatial View)

Show each frequency group's centroid path in 3D weight space over training.

## Cell 9: Cross-Variant Attn PC3 Compression Event

**Metric**: `lead = second_descent_onset_epoch − attn_out_pc3_min_epoch`
- Negative lead = minimum occurs AFTER onset (compression during second descent)
- Positive lead = minimum occurs BEFORE onset (no compression event, or early minimum)
- `MinEp = 0` = minimum is at initialization, no compression event at all

## Cell 10: Within-Group Contribution to Global Trajectory PCA

Different from the spatial PCA in Cell 5 (where the final-epoch W_in neuron positions
determine the basis), here **epochs are points**: PCA is fit on how the flattened
W_in matrix moves through parameter space over the course of training.

For each frequency group, non-member neurons are zeroed out at every epoch and the
masked matrix is projected through the global trajectory basis. This gives each
group's independent **contribution** to the dominant directions of training-time change.

PC3 here = how much each group uses the 3rd direction of global weight-space motion.
If PC3 is working space, groups should expand into it at different times — and those
moments should correspond to the expansion events visible in the activation-space PC3.

```python
import colorsys as _colorsys


def _traj_color(freq_idx: int, n_freq: int) -> str:
    hue = freq_idx / max(n_freq, 1)
    r, g, b = _colorsys.hls_to_rgb(hue, 0.50, 0.65)
    return f"rgb({int(r*255)},{int(g*255)},{int(b*255)})"


def load_win_trajectory(variant, epoch_limit=None):
    """Load W_in at all checkpoints, flattened to one row per epoch.

    Returns (epoch_list, traj_matrix) where traj_matrix is (n_epochs, d_model*d_mlp).
    """
    loader = ArtifactLoader(str(variant.variant_dir / 'artifacts'))
    available = sorted(loader.get_epochs('parameter_snapshot'))
    if epoch_limit is not None:
        available = [e for e in available if e <= epoch_limit]
    rows = []
    for epoch in available:
        snap = loader.load_epoch('parameter_snapshot', epoch)
        rows.append(extract_neuron_weight_matrix(snap).flatten())
    return available, np.stack(rows)


def fit_trajectory_pca(traj_matrix, n_components=3):
    """PCA across epochs (epochs as points).

    Returns (coords, basis, center, var_explained).
    """
    center = traj_matrix.mean(axis=0)
    X      = traj_matrix - center
    _, S, Vt = np.linalg.svd(X, full_matrices=False)
    basis  = Vt[:n_components]
    coords = X @ basis.T
    var_explained = (S**2 / (S**2).sum())[:n_components]
    return coords, basis, center, var_explained


def project_trajectory(traj_matrix, basis, center):
    """Project a trajectory matrix through a pre-fit basis."""
    return (traj_matrix - center) @ basis.T


def load_group_contribution_data(variant):
    """Per-group contribution to the global W_in trajectory PCA.

    For each frequency group: zero out non-member neurons in W_in at every epoch,
    project through the global trajectory basis. This extracts each group's
    independent contribution to the dominant directions of training change.

    Distinct from the spatial centroid approach in Cell 5, where PCA is fit on
    final-epoch neuron positions and groups are represented by centroid coordinates.
    Here epochs are points: the basis captures *how the weight matrix moves*.
    """
    loader = ArtifactLoader(str(variant.variant_dir / 'artifacts'))
    group_data       = loader.load_cross_epoch('neuron_group_pca')
    neuron_group_idx = group_data['neuron_group_idx']  # (d_mlp,)
    group_freqs      = group_data['group_freqs']       # (n_groups,)

    epochs_all, traj_all = load_win_trajectory(variant)
    global_coords, basis, center, var_exp = fit_trajectory_pca(traj_all)

    snap         = loader.load_epoch('parameter_snapshot', epochs_all[0])
    W_in_sample  = extract_neuron_weight_matrix(snap)
    d_model, d_mlp = W_in_sample.shape

    group_coords = {}
    for g_idx in range(len(group_freqs)):
        members = np.where(neuron_group_idx == g_idx)[0]
        if len(members) == 0:
            continue
        col_mask = np.zeros(d_mlp, dtype=bool)
        col_mask[members] = True
        group_traj = traj_all.reshape(-1, d_model, d_mlp).copy()
        group_traj[:, :, ~col_mask] = 0
        group_traj = group_traj.reshape(-1, d_model * d_mlp)
        X_group = group_traj - center
        group_coords[g_idx] = X_group @ basis.T  # (n_epochs, 3)

    return {
        'epochs':        epochs_all,
        'global_coords': global_coords,
        'group_coords':  group_coords,
        'group_freqs':   group_freqs,
        'var_explained': var_exp,
    }


print(f'Loading within-group trajectory contribution data for {WEIGHT_FOCUS}...')
w_focus_variant = family.get_variant(
    prime=weight_focus_v['prime'],
    seed=weight_focus_v['seed'],
    data_seed=weight_focus_v['dseed'],
)
contrib_data = load_group_contribution_data(w_focus_variant)
print(f'  {len(contrib_data["epochs"])} epochs, {len(contrib_data["group_freqs"])} groups')
v = contrib_data['var_explained']
print(f'  Trajectory PCA var: PC1={v[0]:.1%}  PC2={v[1]:.1%}  PC3={v[2]:.1%}')
```

## Cell 11: Shared Timeline — Expansion/Contraction Across Scales

Three signals on a shared epoch axis for a single variant:

1. **Circularity per site** — ring formation (high = ring, low = spreading or reorganizing)
2. **Activation-space PC3 per site** — how far class centroids spread into the 3rd dimension

If grokking is a global reorganization event, these signals should co-move.
If individual components lead or lag, their movements will be offset relative to the
