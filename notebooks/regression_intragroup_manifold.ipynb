```python
_SHAPE_TO_INT = {"flat/blob": 0, "bowl": 1, "saddle": 2}
_INT_TO_SHAPE = {v: k for k, v in _SHAPE_TO_INT.items()}

n_epochs: int = 4
n_groups: int = 3
seed: int = 42
shape_choices = [0, 1, 2]
shape_choices = list(_SHAPE_TO_INT.values())
print(f"shape_choices: {shape_choices}")
rng = np.random.default_rng(seed)
shape_int = rng.choice(shape_choices, (n_epochs, n_groups)).astype(np.int32)

print(f"shape_int")

final_shape_int = shape_int[-1, :]
print(f"final_shape_int: {final_shape_int}")

final_shape = [_INT_TO_SHAPE.get(int(v), "flat/blob") for v in final_shape_int]
print(f"final_shape: {final_shape}")
```
